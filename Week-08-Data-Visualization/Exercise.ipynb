```python
sales_df = pd.read_csv("data/Sales.csv")
```

```python
sales_df.head()
```

```python
sns.lineplot(x = sales_df['month_number'], y = sales_df['facecream'])
sns.lineplot(x = sales_df['month_number'], y = sales_df['facewash'])
sns.lineplot(x = sales_df['month_number'], y = sales_df['toothpaste'])
sns.lineplot(x = sales_df['month_number'], y = sales_df['bathingsoap'])
sns.lineplot(x = sales_df['month_number'], y = sales_df['shampoo'])
sns.lineplot(x = sales_df['month_number'], y = sales_df['moisturizer'])
```

```python
print("They should push bathing soap more and get rid of moisturizer")
```

```python
total_sales = sales_df['facecream'].sum() + sales_df['facewash'].sum() + sales_df['toothpaste'].sum() + sales_df['bathingsoap'].sum() + sales_df['shampoo'].sum() + sales_df['moisturizer'].sum()
print(total_sales)
```

```python
print("Total percentage of sales facecream makes up is: ", (sales_df['facecream'].sum() / total_sales) * 100)
print("Total percentage of sales facewash makes up is: ", (sales_df['facewash'].sum() / total_sales) * 100)
print("Total percentage of sales toothpaste makes up is: ", (sales_df['toothpaste'].sum() / total_sales) * 100)
print("Total percentage of sales bathingsoap makes up is: ", (sales_df['bathingsoap'].sum() / total_sales) * 100)
print("Total percentage of sales shampoo makes up is: ", (sales_df['shampoo'].sum() / total_sales) * 100)
print("Total percentage of sales moisturizer makes up is: ", (sales_df['moisturizer'].sum() / total_sales) * 100)
print("Probably a better way of doing this but :P")
```
